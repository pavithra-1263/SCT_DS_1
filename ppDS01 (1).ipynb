```python
import matplotlib.pyplot as plt
import numpy as np
# Data
age_groups = ['0-4', '5-9', '10-14', '15-19', '20-24', '25-29', '30-34', '35-39', 
              '40-44', '45-49', '50-54', '55-59', '60-64', '65-69', '70-74', '75+']
male_percent = [4.2, 4.1, 3.9, 3.7, 3.6, 3.8, 4.0, 3.9, 3.5, 3.2, 2.8, 2.4, 2.0, 1.6, 1.2, 1.1]
female_percent = [3.9, 3.8, 3.6, 3.5, 3.5, 3.7, 3.9, 3.8, 3.4, 3.1, 2.8, 2.5, 2.2, 1.8, 1.4, 1.5]
total_percent = [8.1, 7.9, 7.5, 7.2, 7.1, 7.5, 7.9, 7.7, 6.9, 6.3, 5.6, 4.9, 4.2, 3.4, 2.6, 2.6]
fig, ax = plt.subplots(figsize=(10, 8))
male = np.array(male_percent)
female = np.array(female_percent)
ax.barh(age_groups, male, color='blue', label='Male')
ax.barh(age_groups, -female, color='pink', label='Female')
ax.set_xlabel('Percentage of Population')
ax.set_ylabel('Age Groups')
ax.set_title('Population Pyramid')
ax.legend()
xticks = np.arange(-5, 6, 1)
ax.set_xticks(xticks)
ax.set_xticklabels(np.abs(xticks))
ax.grid(True, linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()
```
